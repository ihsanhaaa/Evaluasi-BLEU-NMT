```python
import bleu

# Note: the builder functions return things in *corpus* format. 
trans = bleu.load_translation_corpus('/content/drive/MyDrive/Corpus NMT Translasi/Evaluasi BLEU/Target Prediksi Firza.txt') #file prediksi
refs = bleu.load_reference_corpus(['/content/drive/MyDrive/Corpus NMT Translasi/Evaluasi BLEU/Target Uji Firza.txt']) #file referensi

# Data is suitable for corpus level evaluation.
score = bleu.bleu_corpus_level(trans, refs, max_order=2)
print('score Bleu', score)
```
